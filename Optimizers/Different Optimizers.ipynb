```python
import numpy as np
```

```python
x = 4
alpha = 0.1
num_iterations = 10

# SGD optimization
for i in range(num_iterations): 
    gradient = 2 * x
    x = x - alpha * gradient
    print(f"Iteration {i+1}: x = {x}")
    print(f"Iteration {i+1}: gradient = {gradient}")
```

```python
x = 4
alpha = 0.1
num_iterations = 10
beta_1 = 0.9
m = 0

# Adam optimization
for i in range(num_iterations):
    gradient = 2 * x
    m = beta_1 * m + (1 - beta_1) * gradient
    x = x - alpha * m
    print(f"Iteration {i+1}: x = {x}")
    print(f"Iteration {i+1}: gradient = {gradient}")
    print(f"Iteration {i+1}: m = {m}")
```

```python
x = 4
alpha = 0.1
beta_2 = 0.9
epsilon = 1e-7
num_iterations = 10
s = 0

# RMSprop optimization
for i in range(num_iterations):
    gradient = 2 * x
    s = beta_2 * s + (1 - beta_2) * gradient**2
    x = x - alpha / (np.sqrt(s) + epsilon) * gradient
    print(f"Iteration {i+1}: x = {x}")
    print(f"Iteration {i+1}: gradient = {gradient}")
    print(f"Iteration {i+1}: s = {s}")
```

```python
x = 4
alpha = 0.1
epsilon = 1e-7
num_iterations = 10
G = 0

# Adagrad optimization
for i in range(num_iterations):
    gradient = 2 * x
    G = G + gradient**2
    x = x - alpha / (np.sqrt(G) + epsilon) * gradient
    print(f"Iteration {i+1}: x = {x}")
    print(f"Iteration {i+1}: gradient = {gradient}")
    print(f"Iteration {i+1}: G = {G}")
```

```python
x = 4
rho = 0.95
epsilon = 1e-7
num_iterations = 10
G = 0
s = 0

# Adadelta optimization
for i in range(num_iterations):
    gradient = 2 * x
    G = rho * G + (1 - rho) * gradient**2
    delta_x = - np.sqrt(s + epsilon) / np.sqrt(G + epsilon) * gradient
    s = rho * s + (1 - rho) * delta_x**2
    x += delta_x
    print(f"Iteration {i+1}: x = {x}")
    print(f"Iteration {i+1}: gradient = {gradient}")
    print(f"Iteration {i+1}: G = {G}")
    print(f"Iteration {i+1}: delta_x = {delta_x}")
    print(f"Iteration {i+1}: s = {s}")
```
